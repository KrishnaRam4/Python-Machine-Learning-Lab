```python
from sklearn.datasets import load_breast_cancer # loads a buit-in dataset
from sklearn.model_selection import train_test_split # split data into training and testing sets
from sklearn.preprocessing import StandardScaler # scales fetures
from sklearn.svm import SVC # Support Vector Classifier (SVM model)
from sklearn.metrics import accuracy_score # evaluates models performance

# Load breast cancer dataset
X, y = load_breast_cancer(return_X_y=True)
# X -> Features (30 numerical features like tumor radius, texture, etc.)
# y -> Labels (0 = malignant, 1 = benign)

Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3)
# 70% training data
# 30% testing data

scaler = StandardScaler()
Xtr = scaler.fit_transform(Xtr) # training data
Xte = scaler.transform(Xte) # test data (prevents data leakage)

# Scale features
# 1. Linear Kernel
# Draws a straight line (or hyperplane) to separate classes.
# Best for linearly separable data.

# 2. RBF (Radial Basis Function) Kernel
# Default kernel in SVC.
# Handles complex, non-linear relationships.
# Very powerful and commonly used.

# 3. Polynomial Kernel
# Uses polynomial decision boundaries.
# Can model curved relationships.

kernels = ['linear','rbf','poly']
for kernel in kernels:
    svm = SVC(kernel=kernel, random_state=42)
    svm.fit(Xtr, ytr) # train
    pred = svm.predict(Xte) # test
    # Model Evaluation
    print(f"{kernel.upper()} Kernel Accuracy: ", accuracy_score(yte, pred))
```
