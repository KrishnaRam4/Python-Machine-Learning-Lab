```python
# Import math module to use logarithm function
import math

# Import Counter to count label frequencies
# Import defaultdict to group dataset row easily
from collections import Counter, defaultdict

# Function to calculate entropy of a list class labels
def entropy(labels):
    total = len(labels) # Total number of samples
    Counts = Counter(labels) # Count occurrences of each class (Yes/No)
    ent = 0.0 # Initialize entropy value

    # Loop through each class count
    for count in Counts.values():
        p = count / total # Probability of the class
        ent -= p * math.log2(p) # Apply entropy formula: -p * log2(p)
    return ent # Return total entropy

    # Calaucate Yes/No
    # In the dataset
    # Yes = 9
    # No = 5
    # total entropy ~ 0.940 (not pure)
    
# Function to calculate Information Gain for a given attribute
def information_gain(dataset, attribute_index):

    # Extract all class labels (last column in dataset)
    labels = [row[-1] for row in dataset]

    # Calculate entropy of the full dataset
    total_entropy = entropy(labels)

    # Create dictionary to store subsets grouped by attribute value
    subsets = defaultdict(list)

    # Split dataset based on attributes values
    for row in dataset:
        subsets[row[attribute_index]].append(row)

    weighted_entropy = 0.0 # Initialize weighted entropy
    total_size = len(dataset) # Total number of samples

    # Calculate entropy for each subset
    for subset in subsets.values():
        subset_labels = [row[-1] for row in subset] # Extract labels

        # Add weighted entropy of subset
        weighted_entropy += (len(subset) / total_size) * entropy(subset_labels)

    # Information Gain forumla
    ig = total_entropy - weighted_entropy
    return ig # Return Information Gain

# Dataset (Play Tennis dataset)
dataset = [
    ["Sunny", "Hot", "High", "Weak", "No"],
    ["Sunny", "Hot", "High", "Strong", "No"],
    ["Overcast", "Hot", "High", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Strong", "No"],
    ["Overcast", "Cool", "Normal", "Strong", "Yes"],
    ["Sunny", "Mild", "High", "Weak", "No"],
    ["Sunny", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "Normal", "Weak", "Yes"],
    ["Sunny", "Mild", "Normal", "Strong", "Yes"],
    ["Overcast", "Mild", "High", "Strong", "Yes"],
    ["Overcast", "Hot", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Strong", "No"]
]

# Attribute names (excluding target label)
attributes = ["Outlook", "Temperature", "Humidity", "Wind"]

# Function to compute best root node
def compute_root_node():
    print("\n--Information Gain for Each Attribute--")
    gains = {} # Dictionary to store IG values

    # Loop through each attribute
    for i, attr in enumerate(attributes):
        ig = information_gain(dataset, i) # Calculate IG
        gains[attr] = ig # Store IG
        print(f"{attr}: {ig:.4f}") # Print IG rounded to 4 decimals

    # Select attribute with maximum Information Gain
    root = max(gains, key = gains.get)
    print("\n--Best Attributes for Root Gain--")
    print("Root Attribute: ", root)
    return root # Return best attribute

if __name__ == "__main__":
    compute_root_node()

# Attribute | Infomtion Gain | Meaning
#------------------------------------------
# Outlook:          0.2467   Best split
# Temperature:      0.0292   Very weak split
# Humidity:         0.1518   Good split
# Wind:             0.0481   Weak split

# The highest Information Gain is Outlook = 0.2467

        #                 Outlook
        #        ┌──────────┼──────────┐
        #      Sunny      Overcast      Rain
        #        │             │           │
        #    Humidity        Yes        Wind
        #   ┌───────┐                  ┌───────┐
        # High    Normal             Strong    Weak
        #  │         │                 │         │
        # No        Yes               No        Yes

# Best root node attribute is Outlook in Play Tennis dataset
```
