```python
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Load Iris dataset
X = load_iris().data

# Apply GaussianMixture clustering
gmm = GaussianMixture(
    n_components = 3 # Clusters 3
)

# Fits the model to the data.
# Assigns each data point to one of the 3 clusters.
labels = gmm.fit_predict(X)

# Visualize clusters
plt.scatter(
    X[:, 0], # X axis
    X[:, 1], # Y axis
    c = labels # Color
)
plt.title("Gaussian Mixture Model Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Check Convergence
print("Coverged: ", gmm.converged_) # the EM (Expectation-Maximization) algorithm successfully converged.
```
