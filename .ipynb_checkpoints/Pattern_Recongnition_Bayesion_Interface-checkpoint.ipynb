```python
import numpy as np
np.random.seed(42)
class_A = np.random.randn(50,2) + np.array([2,2])
class_B = np.random.randn(50, 2) + np.array([6,6])
print("Class A shape: ", class_A.shape)
print("Class B shape: ", class_B.shape)
prior_A = len(class_A) / (len(class_A) + len(class_B))
prior_B = len(class_B) / (len(class_A) + len(class_B))
print("\nP(Class A) = ", prior_A)
print("P(Class B) = ", prior_B)
mean_A = np.mean(class_A, axis = 0)
mean_B = np.mean(class_B, axis = 0)
cov_A = np.cov(class_A, rowvar = False)
cov_B = np.cov(class_B, rowvar = False)
var_A = np.diag(cov_A)
var_B = np.diag(cov_B)
print("\nMean A: ", mean_A)
print("Mean B: ", mean_B)

def gaussian_likelihood(X, mean, var):
    return np.exp( -0.5 * np.sum((X -mean) ** 2 / var) / np.sqrt(
        (2 * np.pi) ** len(X) * np.prod(var)
    ))

def classify_point(X):
    likelihood_A = gaussian_likelihood(X, mean_A, var_A)
    likelihood_B = gaussian_likelihood(X, mean_B, var_B)
    posterior_A = likelihood_A * prior_A
    posterior_B = likelihood_B * prior_B
    total = posterior_A + posterior_B
    posterior_A /= total
    posterior_B /= total
    return posterior_A, posterior_B

test_point = np.array([4.0, 4.0])
pA, pB = classify_point(test_point)

print("\nTest Point: ", test_point)
print(f"P(Class A | X) = {pA:.4f}")
print(f"P(Class B | X) = {pB:.4f}")
if pA > pB:
    print("Prediction: Class A")
else:
    print("Prediction: Class B")
```
