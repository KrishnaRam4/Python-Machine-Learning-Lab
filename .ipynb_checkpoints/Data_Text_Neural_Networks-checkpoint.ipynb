```python
from sklearn.datasets import fetch_20newsgroups # loads text dataset
from sklearn.feature_extraction.text import TfidfVectorizer # converts text to numerical features
from sklearn.model_selection import train_test_split # splits data into training and testing
from sklearn.neural_network import MLPClassifier # neural network classifier
from sklearn.metrics import accuracy_score # measures model performance

# Load 20 Newsgroups dataset
X, y = fetch_20newsgroups(return_X_y = True)
# X -> raw text posts
# y -> category labels (0–19)
# There are 20 different newsgroup categories like politics, sports, computers, religion, etc

# Convert text to TF-IDF features
X = TfidfVectorizer(
    stop_words='english' # Remove English stopwords(this, is, and, etc.)
).fit_transform(X)
# TF-IDF
# TF (Term Frequency) -> how often a word appears in a documents
# IDF (Inverse Documents Frequency) -> reduces weight of common words 

# Split dataset
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size = 0.3)
# 70% -> training data (learn)
# 30% -> testing data (evaluate performance)

# Create the Neural Network (MLP)
# Multi-Layer Perceptron (MLP) Classifier
model = MLPClassifier(
    hidden_layer_sizes = (100, 50), # 2 hidden layers
    max_iter = 200, # maximum training iterations
    random_state = 42 # ensures reproducibility
)
model.fit(Xtr, ytr)

# Evaluate model
print("Accuracy: ", accuracy_score(yte, model.predict(Xte)))
```
