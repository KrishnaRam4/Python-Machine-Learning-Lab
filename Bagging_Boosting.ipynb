```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

X, y = fetch_california_housing(return_X_y = True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size = 0.3)

bag = BaggingRegressor(DecisionTreeRegressor(), n_estimators = 10)
bag.fit(Xtr, ytr)
bag_pred = bag.predict(Xte)

boost = AdaBoostRegressor(DecisionTreeRegressor(), n_estimators = 10)
boost.fit(Xtr, ytr)
boost_pred = boost.predict(Xte)

print("Bagging MSE: ", mean_squared_error(yte, bag_pred))
print("Bagging R2: ", r2_score(yte, bag_pred))
print("\nBoosting MSE: ", mean_squared_error(yte, boost_pred))
print("Boosting R2: ", r2_score(yte, boost_pred))
```
