```python
from sklearn.datasets import fetch_california_housing # load dataset
from sklearn.model_selection import train_test_split # split data into training/test sets
from sklearn.tree import DecisionTreeRegressor # base model
from sklearn.ensemble import BaggingRegressor, # bagging ensemble model
AdaBoostRegressor # boosting ensemble model
from sklearn.metrics import mean_squared_error, r2_score # evluation metrics

# Load the dataset
X, y = fetch_california_housing(return_X_y = True)
# X -> feature (e.g., modian icome, house age, etc.)
# y -> target (median house value)

Xtr, Xte, ytr, yte = train_test_split(X, y, test_size = 0.3)
# 70 % - training data
# 30 % - testing data

# Bagging Model
bag = BaggingRegressor(
    DecisionTreeRegressor(),
    n_estimators = 10 # build 10 trees
)
bag.fit(Xtr, ytr) # train
bag_pred = bag.predict(Xte) # test

# Boosting Model
boost = AdaBoostRegressor(DecisionTreeRegressor(), n_estimators = 10)
boost.fit(Xtr, ytr)
boost_pred = boost.predict(Xte)

# Model Evalution
print("Bagging MSE: ", mean_squared_error(yte, bag_pred))
print("Bagging R2: ", r2_score(yte, bag_pred))
print("\nBoosting MSE: ", mean_squared_error(yte, boost_pred))
print("Boosting R2: ", r2_score(yte, boost_pred))
```
